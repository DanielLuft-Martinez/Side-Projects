```python
# standard comments and imports
# jsut the basics
import string
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import random
import datetime
import boto3
```

```python
plt.rc('figure', figsize=(40, 40))
```

```python
# api keys

# IEX
IEX_API_KEY = ""
with open('iexcloud.key.txt', 'r') as file:
    IEX_API_KEY = file.readline()

# AWS
AWS_ACCESS_KEY = ''
AWS_SECRET_KEY = ''
with open('aws.rootkey.key.csv', 'r') as file:
    AWS_ACCESS_KEY  =  file.readline()[15:-1]
    AWS_SECRET_KEY = file.readline()[13:]

session = boto3.Session(
aws_access_key_id=AWS_ACCESS_KEY,
aws_secret_access_key=AWS_SECRET_KEY
)
```

```python
tickers_banks = [
            'JPM',
            'BAC',
            'C',
            'WFC',
            'GS',
            'BNPQY'
             ]

# JPMorgan Chase (JPM)
# Bank of America (BAC)
# Citigroup (C)
# Wells Fargo (WFC)
# Goldman Sachs (GS)
# BNP Paribas (BNPQY)
```

```python
ticker_banks_string = ''
for x in tickers_banks:
    ticker_banks_string += x+","
ticker_banks_string = ticker_banks_string[:-1]

# """AGREGATE DESIRED TICKERS"""
ticker_banks_string
```

```python
# creating endpoints
endpoints = 'chart'
years = '1y'
# limited by paywall

# form http request
HTTP_req = f'https://cloud.iexapis.com/stable/stock/market/batch?symbols={ticker_banks_string}&types={endpoints}&range={years}&token={IEX_API_KEY}'
```

```python
bank_data = pd.read_json(HTTP_req)
```

```python
series_dict = {}

for ticker in tickers_banks:
    series_dict.update({ticker : pd.DataFrame(bank_data[ticker]['chart'])['close']})
    
series_list = [pd.DataFrame(bank_data[ticker]['chart'])['close'] for ticker in tickers_banks]

series_list.append(pd.DataFrame(bank_data[ticker]['chart'])['date'])
                            
col_names = tickers_banks.copy()
col_names.append('Date')
        
bank_data_concat = bank_data.copy()
bank_data_concat = pd.concat(series_list, axis=1)
                            
bank_data_concat.columns = col_names
  
bank_data_concat
bank_data_concat.set_index('Date', inplace = True)
```

```python
%matplotlib inline
```

```python
fig = plt.figure()


ax = fig.add_subplot(111)

ax.boxplot(bank_data_concat.T)

plt.title('Boxplot of Bank Stock Prices (1 year lookback)', fontsize = 50)

ticks = range(1, len(bank_data.columns)+1)
labels = list(bank_data.columns)




plt.xticks(ticks,labels, fontsize = 50)
plt.yticks(fontsize = 50)





ax.plot([1],[1])
ax.tick_params(top='off', bottom='off', left='off', right='off', labelleft='on', labelbottom='on')

today = datetime.datetime.now()

fig.savefig(f'bank_data_imgs/boxplot-{str(today.date())}.png')
```

```python
dates_full = bank_data_concat.index.to_series()
dates_full = [str(pd.to_datetime(d)) for d in dates_full]

```

```python
dates = [ x[5:-9]+'-'+x[2:4] for x in dates_full]
range_points = []
month = ''
for i,d in enumerate(dates):
    if month == '':
        month = d[0:2]
        range_points.append(i)
    if month != d[0:2]:
        month = d[0:2]
        range_points.append(i)

        
```

```python


for ticker in tickers_banks:
    stock_prices = bank_data_concat[ticker]    

    #Add titles to the chart and axes
    plt.title(f'Bank Stock Pricees (1Y Lookback)', fontsize=50)
    plt.ylabel("Stock Price", fontsize=50)
    plt.xlabel("Date", fontsize=50)
    plt.xticks(range_points[1:], fontsize = 35)
    plt.yticks(fontsize=45)
   

    #Generate the scatterplot
    plt.plot(dates, stock_prices, linewidth=7.0) 
  
    
    plt.legend(tickers_banks,fontsize = 40)
    
    
today = datetime.datetime.now()

plt.savefig(f'bank_data_imgs/lineplot-{str(today.date())}.png')
```

```python
plt.hist(bank_data_concat.transpose(), bins = 50, rwidth=1)
plt.legend(bank_data.columns,fontsize=50)
plt.title("A Histogram of Daily Closing Bank Stock Prices (1Y Lookback)", fontsize = 50)
plt.ylabel("Observations", fontsize = 50)
plt.xlabel("Stock Prices", fontsize = 50)
plt.xticks(fontsize = 45)
plt.yticks(fontsize=45)


today = datetime.datetime.now()

plt.savefig(f'bank_data_imgs/histogram-{str(today.date())}.png')
```
